```python
import pandas as pd
from collections import defaultdict
```

# CSV

```python
df = pd.read_csv('sample_data/sample_transactions.csv')
df.head()
```

```python
user_total = defaultdict(float)

for index, row in df.iterrows():
    user_total[row["customer_name"]] += row["amount"]
```

```python
print(user_total.items())
```

# JSONL

```python
df = pd.read_json('sample_data/sample_events.jsonl', lines=True)
df['date'] = pd.to_datetime(df['timestamp']).dt.normalize()

df.head()
```

```python
daily_event_counts = defaultdict(int)

for index, row in df.iterrows():
    daily_event_counts[row["date"]] += 1

print(daily_event_counts.items())
```
