```python
API_ENDPOINT = "https://api.coingecko.com/api/v3/coins/markets"
API_PARAMETERS = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 3,
    "sparkline": "false"
}

#API_URL = "https://api.github.com/orgs/apache/repos?per_page=50&page=1"
API_URL = "https://rickandmortyapi.com/api/character?page=1"
```

```python
import requests
import pandas as pd
import time
```

```python
def extract_paginated_data(base_url, params=None):
    all_data = []
    page = 1
    max_pages = 10
    has_more = True

    while has_more:
        current_params = params.copy() if params else {}
        current_params['page'] = page

        response = requests.get(base_url, params=current_params)
        response.raise_for_status()

        if response.status_code == 429:
            retry_after = int(response.headers.get("Retry-After", 8))
            print(f"Rate limited, sleeping for {retry_after}")
            time.sleep(retry_after)
            continue

        data = response.json()
        if not data or page > max_pages:
            has_more = False
        else:
            all_data.extend(data['results'])
            page += 1

    return all_data
```
