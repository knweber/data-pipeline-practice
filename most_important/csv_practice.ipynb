```python
import csv
```

```python
with open("sample_data/sample_transactions.csv", mode="r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
with open("output/transactions.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["CustomerID", "CustomerName"])
    writer.writerow(["123", "Sally"])
```

```python
# pandas - process large CSV files using chunksize
```

```python
import pandas as pd

is_first_chunk = True

for chunk in pd.read_csv("sample_data/sample_transactions.csv", chunksize=1):
    chunk.to_csv(
        "output/pandas_chunksize_output.csv",
        mode='a',
        header=is_first_chunk
    )

    is_first_chunk = False
```
