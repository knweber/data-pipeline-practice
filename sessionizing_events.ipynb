```python
#sample data
events = [
    {"user": "A", "time": "2026-07-09 10:00"},
    {"user": "A", "time": "2026-07-09 10:05"},
    {"user": "A", "time": "2026-07-09 10:45"},
    {"user": "B", "time": "2026-07-09 11:00"},
    {"user": "B", "time": "2026-07-09 11:10"},
]
```

```python
from collections import defaultdict
from datetime import datetime, timedelta
import pandas as pd
```

```python
#group events by user
grouped = defaultdict(list)

for event in events:
  grouped[event["user"]].append(event) # all events by user

print(grouped)
```

```python
# sort events by user by time
output = []

for user, user_events in grouped.items():
    user_events.sort(key=lambda e: event["time"])

    session = 1
    previous = None

    for event in user_events:
        # if there is a previous event and if it's been more than 30 minutes since, create a new session
        if previous and event["time"] - previous > timedelta(minutes=30):
            session += 1

        output.append({
            "user": user,
            "time": event["time"],
            "session": session
        })

        previous = event["time"]


print(output)
```

```python
session_counts = defaultdict(int)

for session in output:
    session_counts[session['session']] += 1

print(session_counts)
```
